```python
import re
import string
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

res_str = open("../data/manual-scraping.txt").read().lower()

# remove number
res_str = re.sub(r'\d+', '', res_str)

# remove special character
res_str = res_str.translate(str.maketrans('','',string.punctuation))

# remove white space
res_str = res_str.strip()

res_str
```

```python
# TOKENIZE
import nltk
from nltk.probability import FreqDist

list_token = nltk.tokenize.word_tokenize(res_str)

# remove one char token
list_token = filter(lambda x:len(x)!=1,list_token)

list_token = list(list_token)
list_token
# list_token = list(list_token)


```

```python
# FILTERING
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

list_stopwords = open("../data/stopword-list.txt").read().lower()
list_stopwords = nltk.tokenize.word_tokenize(list_stopwords)

# len(list_stopwords)
# len(list_token)

res_list_token = []

for token in list_token:
  if any (substring in token for substring in list_stopwords):
    continue
  else:
    res_list_token.append(token)  

res_list_token

freq_token = FreqDist(res_list_token).most_common()

freq_token
```
