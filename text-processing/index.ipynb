```python
import re
import string
import pandas as pd
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
# create stemmer
factory = StemmerFactory()
stemmer = factory.create_stemmer()

# res_str = open("../data/Muhammad-Ridwan-123.csv").read().lower()

df = pd.read_csv("../data/scraping/Muhammad-Ridwan-123.csv")
df[['answer']].iloc[2].values[0]


def remove_punctuation(sentence):
    # remove number
    result = re.sub(r'\d+', '', sentence)

    # remove special character
    result = re.sub(r'[^a-zA-Z]'," ",result)
    return result


answers_filtered_punctuation = []
questions_filtered_punctuation = []

for index,row in df.iterrows():
    answers_filtered_punctuation.append(stemmer.stem(remove_punctuation(row['answer']).lower().strip()))
    questions_filtered_punctuation.append(stemmer.stem(remove_punctuation(row['question'].lower().strip())))

df_filtered_punctutation = pd.DataFrame({'questions':questions_filtered_punctuation,'answers':answers_filtered_punctuation})
df_filtered_punctutation

```

```python
# TOKENIZE
import nltk
from itertools import chain
from nltk.probability import FreqDist

list_answers = df_filtered_punctutation['answers'].values.tolist()
list_questions = df_filtered_punctutation['questions'].values.tolist()


def remove_one_char_token(list_token):
    res = filter(lambda x:len(x)!=1,list_token)
    return list(res)

# ['i love you','i miss you'] => [['i','love','you'],['i','miss','you']]
def tokenize_list_of_sentence(list_sentence):
    res = []
    for sentence in list_sentence:
        res.append(remove_one_char_token(nltk.tokenize.word_tokenize(sentence)))
    return res

dict_ans_quest = {
    "answers":list(chain.from_iterable (tokenize_list_of_sentence(list_answers))),
    "questions":list(chain.from_iterable(tokenize_list_of_sentence(list_questions)))
}

print(dict_ans_quest)



```

```python
# FILTERING (remove stopwords)
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

factory = StopWordRemoverFactory()
stopword = factory.create_stop_word_remover()

list_stopwords = open("../data/stopword-list.txt").read().lower()
list_stopwords = nltk.tokenize.word_tokenize(list_stopwords)


def get_freq_token(list_token):
    res = []
    for token in list_token:
      if any (substring in token for substring in list_stopwords):
        continue
      else:
        res.append(token)  
    freq_token = FreqDist(res).most_common()
    return freq_token

freq_dict_ans_quest = {
    "answers":get_freq_token (dict_ans_quest["answers"]),
    "questions":get_freq_token (dict_ans_quest["questions"])
}
freq_dict_ans_quest


```
